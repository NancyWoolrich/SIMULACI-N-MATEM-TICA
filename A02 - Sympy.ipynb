**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x= sp.var("x")

def f(x):
    return 1260+117*x-9*x**2

dy=sp.diff(f(x),x)
d2y=sp.diff(dy,x)
```

```python
pc=sp.solve(dy,x)
opt=f(pc[0])
pc, opt
```

```python
d2y.subs(x,pc[0])
```

Es máximo -> ¿Local o global?

```python
x_n=np.linspace(-15,20,1000)
y_n=f(x_n)
plt.figure() 
plt.plot(x_n, y_n)
plt.grid()
plt.plot(13/2,6561/4,"*r",ms=10)
```

**Es máximo global.**

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=sp.var("x")

def j(x):
    return (1/4*x**4)+(1/3*x**3)-x**2

dy=sp.diff(j(x),x)
d2y=sp.diff(dy,x)
```

```python
pc=sp.solve (dy,x)
valores_optimos=[]
for p in pc:
    par_ordenado=[p,j(p)]
    valores_optimos.append(par_ordenado)
valores_optimos
```

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

1. Definir criterio de optimización
 Maximizar volumen
 
2. Revisar qué variables entran en juego
$ l $ - lado
$ b $ - base
$ r $ - radio
$ h $ - altura
$ v $ - volumen

3. Definir función de costo

$$ V=\frac{1}{3}\pi r^2h $$

4. información adicional

$$ b+2l=10 $$
$$ r=\frac{b}{2} $$
$$ l^2=r^2+h^2 $$



```python
import sympy as sp

r=sp.var("r")
h=sp.var("h")
l=sp.var("l")
b=sp.var("b")

def V(r,h):
    return sp.pi*r**2*h/3
V(r,h)
```

```python
restriccion=b+2*l-10
radio=b/2
pitagoras=l**2-r**2-h**2
```

```python
altura=sp.solve(pitagoras,h)[1]
altura
```

```python
V(r,h)
```

```python
V(r,h).subs(h,altura).subs(r,radio)
```

```python
l_sol=sp.solve(restriccion,l)[0]
l_sol
```

```python
V_b=V(r,h).subs(h,altura).subs(r,radio).subs(l,l_sol)
V_b
```

```python
dV=sp.diff(V_b,b)
dV
```

```python
sp.solve(dV,b)
```

La base del triángulo debe ser 4 cm.

```python
l_sol.subs(b,4)
```

Los lados del triángulo deben medir 3 cm cada uno.

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

```python
import sympy as sp

x=sp.var("x")
y=sp.var("y")

def A(x,y):
    return x*y

restriccion=x+2*y-6
```

```python
x_y=sp.solve(restriccion,x)[0]
x_y
```

```python
A_y=A(x,y).subs(x,x_y)
```

```python
dA=sp.diff(A_y,y)
dA
```

```python
sp.solve(dA,y)[0]
```

Los postes deben medir 1.5 m.

```python
x_y.subs(y,1.5)
```

El travesaño debe medir 3 m.
